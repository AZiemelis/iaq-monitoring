# Temperature from 2023-10-24

```python
# SQL query to select relevant temperature data
sql_query = """
SELECT s.name, sd.time, sd.temperature
FROM sensor_data sd
JOIN sensors s ON sd.sensor = s.id
WHERE CAST(sd.time AS time) BETWEEN '08:00:00' AND '22:00:00'
AND sd.time::date >= '2023-10-24'
"""
# Execute the SQL query and fetch the data into a DataFrame
df = pd.read_sql_query(sql_query, engine)
```

```python
# Sensor names you want to analyze in detail
specific_sensors = ['C2206']  # Replace these with your actual sensor names

# Initialize a dictionary to count occurrences for each sensor
occurrences_by_name = {name: 0 for name in df['name'].unique()}
# Initialize a list to store detailed occurrences for specific sensors, now including average temperature
specific_sensor_occurrences = []

# Iterate through each sensor's data
for name, group in df.groupby('name'):
    # Ensure the group is sorted by time to correctly calculate periods
    group = group.sort_values(by='time')
    temperature_condition = (group['temperature'] < 20) | (group['temperature'] > 26)
    in_period = False
    period_start_index = None

    for i in range(len(group)):
        # Extract time and date from timestamp for comparison
        current_time = group['time'].iloc[i].time()
        current_date = group['time'].iloc[i].date()
        # Start a new period if the condition is met and before 22:00
        if temperature_condition.iloc[i] and current_time < pd.to_datetime('22:00:00').time():
            if not in_period:
                period_start_index = i
                in_period = True
            # Determine if it's the last record, the next condition is not met, the next time exceeds 22:00, or the date changes
            is_last_record = i == len(group) - 1
            next_condition_not_met = i < len(group) - 1 and not temperature_condition.iloc[i + 1]
            next_time_exceeds = i < len(group) - 1 and group['time'].iloc[i + 1].time() >= pd.to_datetime('22:00:00').time()
            date_changes = i < len(group) - 1 and group['time'].iloc[i + 1].date() != current_date
            if is_last_record or next_condition_not_met or next_time_exceeds or date_changes:
                if (group['time'].iloc[i] - group['time'].iloc[period_start_index]).total_seconds() > 3600:
                    occurrences_by_name[name] += 1
                    if name in specific_sensors:
                        start_time = group['time'].iloc[period_start_index]
                        end_time = group['time'].iloc[i]
                        duration = end_time - start_time
                        # Calculate the average temperature over the period
                        average_temperature = group['temperature'].iloc[period_start_index:i+1].mean()
                        specific_sensor_occurrences.append((name, start_time, end_time, duration, average_temperature))
                in_period = False
        else:
            in_period = False

# Create a DataFrame from the occurrences dictionary for all sensors
occurrences_df = pd.DataFrame(list(occurrences_by_name.items()), columns=['Sensor Name', 'Occurrences'])

# Create a DataFrame for specific sensor occurrences, now including average temperature
specific_occurrences_df = pd.DataFrame(specific_sensor_occurrences, columns=['Sensor Name', 'Start Time', 'End Time', 'Duration', 'Average Temperature'])

specific_occurrences_df
```

```python
sql_query = """
SELECT s.name, sd.time, sd.co2
FROM sensor_data sd
JOIN sensors s ON sd.sensor = s.id
"""
# Execute the SQL query and fetch the data into a DataFrame
df = pd.read_sql_query(sql_query, engine)


def analyze_sensor_co2_data(df, start_date, end_date, specific_sensor_name='C4410'):
    """
    Analyze sensor data within a given date range, count occurrences of CO2 levels above 1000,
    and provide detailed occurrences for a specific sensor.

    Parameters:
    - df: DataFrame containing sensor data with columns ['time', 'name', 'co2']
    - start_date: Start of the date range (inclusive)
    - end_date: End of the date range (inclusive)
    - specific_sensor_name: Name of the sensor to analyze in detail (default: 'C4410')
    """

    
    # Convert start_date and end_date to datetime
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    
    
    
    df = df[(df['time'] >= start_date) & (df['time'] <= end_date)]


    
    # Initialize a dictionary with all sensor names to count occurrences
    occurrences_by_name = {name: 0 for name in df['name'].unique()}
    # Dictionary to store occurrences details for a specific sensor
    specific_sensor_occurrences = []

    # Iterate through each sensor name's data
    for name, group in df.groupby('name'):
        above_1000 = group['co2'] > 1000
        in_period = False
        period_start_index = None
        
        for i in range(1, len(group)):
            if above_1000.iloc[i] and above_1000.iloc[i - 1]:
                if not in_period:
                    period_start_index = i - 1
                    in_period = True
                if in_period and (group['time'].iloc[i] - group['time'].iloc[period_start_index]).total_seconds() > 3600:
                    if i < len(group) - 1 and group['co2'].iloc[i + 1] < 1000:
                        occurrences_by_name[name] += 1
                        # Check if this is the specific sensor we're interested in
                        if name == specific_sensor_name:
                            start_time = group['time'].iloc[period_start_index]
                            end_time = group['time'].iloc[i]
                            duration = end_time - start_time
                            specific_sensor_occurrences.append((start_time, end_time, duration))
                        in_period = False
            else:
                in_period = False

    # Convert the general occurrences dictionary to a DataFrame
    occurrences_df = pd.DataFrame(list(occurrences_by_name.items()), columns=['Sensor Name', 'Occurrences'])

    # Add or ensure the 'Floor' column exists, derived from the sensor name
    occurrences_df['Floor'] = occurrences_df['Sensor Name'].apply(lambda x: int(x[1]))

    # Filter the DataFrame to include only sensors with occurrences larger than 2
    filtered_df = occurrences_df[occurrences_df['Occurrences'] > 2]

    # Sort the filtered DataFrame by 'Floor' in ascending order, then by 'Occurrences' in descending order
    sorted_df = filtered_df.sort_values(by=['Floor', 'Occurrences'], ascending=[True, False])

    # Plot
    plt.figure(figsize=(20, 8))  # Adjust size as needed
    ax = sns.barplot(x='Sensor Name', y='Occurrences', hue='Floor', dodge=False, data=sorted_df, palette='hls')
    plt.xticks(rotation=90)  # Rotate sensor names for better visibility
    plt.title(f'Sensors with More Than 2 Occurrences of CO2 > 1000 by Floor ({start_date} to {end_date})')
    plt.ylabel('Number of Occurrences')
    plt.xlabel('Sensor Name')
    plt.legend(title='Floor', bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # Add bar labels
    for container in ax.containers:
        ax.bar_label(container, padding=3)  # padding=3 for a little space between the bar and label

    plt.tight_layout()
    plt.show()


    # Detailed occurrences for the specific sensor, if any
    specific_occurrences_df = pd.DataFrame(specific_sensor_occurrences, columns=['Start Time', 'End Time', 'Duration'])
    if not specific_occurrences_df.empty:
        print(f"\nDetailed Occurrences for Sensor {specific_sensor_name}:")
        print(specific_occurrences_df)
        print(f"\nCount of Occurrences Over 1 Hour for Sensor {specific_sensor_name}: {len(specific_occurrences_df)}")
    else:
        print(f"\nNo detailed occurrences found for Sensor {specific_sensor_name} within the specified date range.")
```
